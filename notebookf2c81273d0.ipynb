```python
# âœ… Imports
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os
import matplotlib.pyplot as plt
import torchmetrics
from tqdm import tqdm
import torchvision.models as models

# âœ… Device
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# âœ… Define Paths
DATASET_PATH = "/kaggle/input/mydata/mother_dataset"
TRAIN_BLUR_DIR = os.path.join(DATASET_PATH, "train", "blur")
TRAIN_SHARP_DIR = os.path.join(DATASET_PATH, "train", "sharp")
VAL_BLUR_DIR = os.path.join(DATASET_PATH, "validation", "blur")
VAL_SHARP_DIR = os.path.join(DATASET_PATH, "validation", "sharp")

# âœ… Transformations
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomCrop((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.RandomRotation(10),
    transforms.GaussianBlur(kernel_size=3),
    transforms.ToTensor(),
])

# âœ… Custom Dataset Class
class DeblurDataset(Dataset):
    def __init__(self, blur_dir, sharp_dir, num_samples, offset=0):
        self.blur_files = sorted(os.listdir(blur_dir))[offset:offset+num_samples]
        self.sharp_files = sorted(os.listdir(sharp_dir))[offset:offset+num_samples]
        self.blur_dir = blur_dir
        self.sharp_dir = sharp_dir

    def __len__(self):
        return len(self.blur_files)

    def __getitem__(self, idx):
        blur_img = Image.open(os.path.join(self.blur_dir, self.blur_files[idx])).convert("RGB")
        sharp_img = Image.open(os.path.join(self.sharp_dir, self.sharp_files[idx])).convert("RGB")
        return transform(blur_img), transform(sharp_img)

# âœ… Data Loaders
train_dataset = DeblurDataset(TRAIN_BLUR_DIR, TRAIN_SHARP_DIR, num_samples=10000, offset=0)
val_dataset = DeblurDataset(VAL_BLUR_DIR, VAL_SHARP_DIR, num_samples=1000)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=2)

# âœ… Define MPRNet Model
class MPRNet(nn.Module):
    def __init__(self):
        super(MPRNet, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv2d(128, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 3, 3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# âœ… Initialize Model
model = MPRNet().to(DEVICE)

# âœ… Loss Functions
criterion_L1 = nn.L1Loss()

class PerceptualLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.vgg = models.vgg19(weights="IMAGENET1K_V1").features[:16].eval().to(DEVICE)
        for param in self.vgg.parameters():
            param.requires_grad = False
        self.criterion = nn.L1Loss()

    def forward(self, x, y):
        return self.criterion(self.vgg(x), self.vgg(y))

criterion_perceptual = PerceptualLoss()

# âœ… Optimizer & Scheduler
optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10, eta_min=1e-6)

# âœ… Metrics
psnr_metric = torchmetrics.image.PeakSignalNoiseRatio().to(DEVICE)
ssim_metric = torchmetrics.image.StructuralSimilarityIndexMeasure().to(DEVICE)

# âœ… Define Epochs & Save Path
EPOCHS = 20
SAVE_PATH = "/kaggle/working/mprnet_phase1_fixed.pth"

best_loss = float("inf")
loss_history, psnr_history, ssim_history = [], [], []

# âœ… Training Loop
for epoch in range(1, EPOCHS + 1):
    model.train()
    running_loss = 0.0
    loop = tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS}")

    for blur_img, sharp_img in loop:
        blur_img, sharp_img = blur_img.to(DEVICE), sharp_img.to(DEVICE)
        optimizer.zero_grad()
        output = model(blur_img)

        loss = 0.6 * criterion_L1(output, sharp_img) + 0.4 * criterion_perceptual(output, sharp_img)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    avg_loss = running_loss / len(train_loader)
    loss_history.append(avg_loss)

    # âœ… Validation
    model.eval()
    with torch.no_grad():
        psnr_avg, ssim_avg = 0.0, 0.0
        for blur_img, sharp_img in val_loader:
            blur_img, sharp_img = blur_img.to(DEVICE), sharp_img.to(DEVICE)
            output = model(blur_img)
            psnr_avg += psnr_metric(output, sharp_img).item()
            ssim_avg += ssim_metric(output, sharp_img).item()
        psnr_avg /= len(val_loader)
        ssim_avg /= len(val_loader)

    psnr_history.append(psnr_avg)
    ssim_history.append(ssim_avg)

    print(f"Epoch [{epoch}/{EPOCHS}] - Loss: {avg_loss:.6f}, PSNR: {psnr_avg:.2f}, SSIM: {ssim_avg:.4f}")
    scheduler.step()

    # âœ… Save Best Model
    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model.state_dict(), SAVE_PATH)
        print(f"âœ… Best Model Saved at Epoch {epoch} with Loss: {best_loss:.6f}")

# âœ… Save Training History
torch.save({"loss": loss_history, "psnr": psnr_history, "ssim": ssim_history}, SAVE_PATH.replace(".pth", "_history.pth"))
print("ðŸŽ¯ Training Complete!")

# âœ… Plot Metrics
plt.figure(figsize=(15,4))
plt.subplot(131); plt.plot(loss_history, '-o', color='red'); plt.title('Loss vs Epochs'); plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.subplot(132); plt.plot(psnr_history, '-s', color='blue'); plt.title('PSNR vs Epochs'); plt.xlabel('Epoch'); plt.ylabel('PSNR')
plt.subplot(133); plt.plot(ssim_history, '-d', color='green'); plt.title('SSIM vs Epochs'); plt.xlabel('Epoch'); plt.ylabel('SSIM')
plt.tight_layout(); plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os
import matplotlib.pyplot as plt
import torchmetrics
from tqdm import tqdm
import torchvision.models as models

# âœ… Paths
DATASET_PATH = "/kaggle/input/mydata/mother_dataset"
TRAIN_BLUR_DIR = os.path.join(DATASET_PATH, "train", "blur")
TRAIN_SHARP_DIR = os.path.join(DATASET_PATH, "train", "sharp")
VAL_BLUR_DIR = os.path.join(DATASET_PATH, "validation", "blur")
VAL_SHARP_DIR = os.path.join(DATASET_PATH, "validation", "sharp")

# âœ… Transformations (Enhanced Augmentation)
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomCrop((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.3, contrast=0.3),  # Increased Jitter
    transforms.RandomRotation(15),  # Increased Rotation for variation
    transforms.RandomApply([transforms.GaussianBlur(kernel_size=3)], p=0.5),
    transforms.RandomApply([transforms.RandomAdjustSharpness(sharpness_factor=2)], p=0.3),
    transforms.ToTensor(),
])

# âœ… Custom Dataset Class (Phase 2 - Remaining 50%)
class DeblurDataset(Dataset):
    def __init__(self, blur_dir, sharp_dir, num_samples, offset=0):
        self.blur_files = sorted(os.listdir(blur_dir))[offset:offset+num_samples]
        self.sharp_files = sorted(os.listdir(sharp_dir))[offset:offset+num_samples]
        self.blur_dir = blur_dir
        self.sharp_dir = sharp_dir

    def __len__(self):
        return len(self.blur_files)

    def __getitem__(self, idx):
        blur_img = Image.open(os.path.join(self.blur_dir, self.blur_files[idx])).convert("RGB")
        sharp_img = Image.open(os.path.join(self.sharp_dir, self.sharp_files[idx])).convert("RGB")
        return transform(blur_img), transform(sharp_img)

# âœ… Load Dataset (Phase 2 - Remaining 50%)
train_dataset_phase2 = DeblurDataset(TRAIN_BLUR_DIR, TRAIN_SHARP_DIR, num_samples=10000, offset=10000)
val_dataset = DeblurDataset(VAL_BLUR_DIR, VAL_SHARP_DIR, num_samples=1000)

train_loader_phase2 = DataLoader(train_dataset_phase2, batch_size=8, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=2)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# âœ… Load Phase 1 Trained Model
class MPRNet(nn.Module):
    def __init__(self):
        super(MPRNet, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 3, kernel_size=3, stride=1, padding=1)
        )
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

model = MPRNet().to(DEVICE)
model.load_state_dict(torch.load("/kaggle/working/mprnet_phase1_fixed.pth"))  # Load Phase 1 Model

# âœ… Loss Functions
criterion_L1 = nn.L1Loss()

class PerceptualLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.vgg = models.vgg19(weights="IMAGENET1K_V1").features[:16].eval().to(DEVICE)
        for param in self.vgg.parameters():
            param.requires_grad = False
        self.criterion = nn.L1Loss()

    def forward(self, x, y):
        return self.criterion(self.vgg(x), self.vgg(y))

criterion_perceptual = PerceptualLoss()

# âœ… Optimizer and Scheduler
optimizer = optim.Adam(model.parameters(), lr=5e-5)  # Fine-tuned LR
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=15, eta_min=1e-6)

# âœ… Metrics
psnr_metric = torchmetrics.image.PeakSignalNoiseRatio().to(DEVICE)
ssim_metric = torchmetrics.image.StructuralSimilarityIndexMeasure().to(DEVICE)

# âœ… Define Save Path
SAVE_PATH = "/kaggle/working/mprnet_final.pth"

best_loss = float("inf")
loss_history, psnr_history, ssim_history = [], [], []

# âœ… Training Loop (Phase 2)
EPOCHS = 30
for epoch in range(1, EPOCHS + 1):
    model.train()
    running_loss = 0.0
    loop = tqdm(train_loader_phase2, desc=f"Epoch {epoch}/{EPOCHS}", leave=True)
    
    for blur_img, sharp_img in loop:
        blur_img, sharp_img = blur_img.to(DEVICE), sharp_img.to(DEVICE)
        optimizer.zero_grad()
        output = model(blur_img)

        loss = 0.5 * criterion_L1(output, sharp_img) + 0.5 * criterion_perceptual(output, sharp_img)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    avg_loss = running_loss / len(train_loader_phase2)
    loss_history.append(avg_loss)

    model.eval()
    with torch.no_grad():
        psnr_avg, ssim_avg = 0.0, 0.0
        for blur_img, sharp_img in val_loader:
            blur_img, sharp_img = blur_img.to(DEVICE), sharp_img.to(DEVICE)
            output = model(blur_img)
            psnr_avg += psnr_metric(output, sharp_img).item()
            ssim_avg += ssim_metric(output, sharp_img).item()
        psnr_avg /= len(val_loader)
        ssim_avg /= len(val_loader)
    psnr_history.append(psnr_avg)
    ssim_history.append(ssim_avg)
    
    print(f"Epoch [{epoch}/{EPOCHS}] - Loss: {avg_loss:.6f}, PSNR: {psnr_avg:.2f}, SSIM: {ssim_avg:.4f}")
    scheduler.step()

    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model.state_dict(), SAVE_PATH)
        print(f"âœ… Best Model Saved at Epoch {epoch} with Loss: {best_loss:.6f}")

print("ðŸŽ¯ Phase 2 Training Complete!")
```
